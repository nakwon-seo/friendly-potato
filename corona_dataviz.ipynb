```python
plt.figure(figsize=(30, 10))

plt.subplot(221)
plt.plot(interesting_countries_df.groupby(['date','location']).sum()['total_cases'].unstack())
plt.title("total cases")
plt.legend(interesting_countries_df["location"].unique())
plt.xticks(rotation='vertical')
plt.subplot(222)
plt.plot(interesting_countries_df.groupby(['date','location']).sum()['new_cases'].unstack())
plt.title("new cases")
plt.legend(interesting_countries_df["location"].unique())
plt.xticks(rotation='vertical')
plt.subplot(223)
plt.plot(interesting_countries_df.groupby(['date','location']).sum()['total_deaths'].unstack())
plt.title("total deaths")
plt.legend(interesting_countries_df["location"].unique())
plt.xticks(rotation='vertical')
plt.subplot(224)
plt.plot(interesting_countries_df.groupby(['date','location']).sum()['new_deaths'].unstack())
plt.title("new deaths")
plt.legend(interesting_countries_df["location"].unique())
#plt.xticks(rotation='vertical')
```
