```python
df_gdp.drop(['Arab World','Central Europe and the Baltics','East Asia & Pacific (excluding high income)','Early-demographic dividend','East Asia & Pacific','Europe & Central Asia (excluding high income)','Europe & Central Asia','Euro area','European Union','Fragile and conflict affected situations','High income','Heavily indebted poor countries (HIPC)','IBRD only','IDA & IBRD total','IDA total','IDA blend','IDA only','Latin America & Caribbean (excluding high income)','Lao PDR','Latin America & Caribbean','Least developed countries: UN classification','Low income','Lower middle income','Low & middle income','Late-demographic dividend','Middle East & North Africa','Middle income','Middle East & North Africa (excluding high income)','OECD members','Other small states','Pre-demographic dividend','West Bank and Gaza','Pacific island small states','Post-demographic dividend','South Asia','Sub-Saharan Africa (excluding high income)','Small states','Turks and Caicos Islands','East Asia & Pacific (IDA & IBRD countries)','Europe & Central Asia (IDA & IBRD countries)','Latin America & the Caribbean (IDA & IBRD countries)','Middle East & North Africa (IDA & IBRD countries)','South Asia (IDA & IBRD)','Sub-Saharan Africa (IDA & IBRD countries)','Upper middle income','World','North America'], axis=0, inplace=True)
df_gdp.head()
```

```python
df_gdp.sort_values(by='Total',ascending=False,axis=0, inplace=True)
df_gdp_top5=df_gdp.head(5)

df_gdp_top5=df_gdp_top5[years].transpose()

print(df_gdp_top5)
```

```python
df_gdp_top5.index=df_gdp_top5.index.map(int)
df_gdp_top5.plot(kind='line', figsize=(16,10))
plt.title('GDP of Top 5 Countries')
plt.ylabel('GDP(current US$)')
plt.xlabel('Years')
plt.show
```

```python
df_gdp_top5.index=df_gdp_top5.index.map(int)
df_gdp_top5.plot(kind='area', 
                 stacked=False,
                 figsize=(20, 10),
                 )
plt.title('GDP of Top 5 Countries')
plt.ylabel('GDP(current US$)')
plt.xlabel('Years')
plt.show
```

```python
df_gdp_top5.index=df_gdp_top5.index.map(int)
df_gdp_top5.plot(kind='area',
                stacked=False,
                alpha=0.3,
                figsize=(20,10),
                )
plt.title('GDP of Top 5 Countries')
plt.ylabel('GDP(current US$)')
plt.xlabel('Years')
plt.show
```

```python
ax = df_gdp_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('GDP of Top 5 Countries')
ax.set_ylabel('GDP(current US$)')
ax.set_xlabel('Years')
```
