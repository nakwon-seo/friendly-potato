```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot'])
```

```python
df_gdp=pd.read_excel("http://api.worldbank.org/v2/en/indicator/NY.GDP.MKTP.CD?downloadformat=excel", skiprows=range(3))
```

```python
df_gdp.shape
```

```python
df_gdp.drop(['Country Code','Indicator Name','Indicator Code'], axis=1, inplace=True)
```

```python
#df_gdp.info()
```

```python
years=list(map(str,range(1980,2019)))
```

```python
#print(type(df_gdp.columns))
#print(type(df_gdp.index))
```

```python
#df_gdp.columns.tolist()
#df_gdp.index.tolist()

#print (type(df_gdp.columns.tolist()))
#print (type(df_gdp.index.tolist()))
```

```python
#df_gdp.shape
```

```python
df_gdp = df_gdp.drop(df_gdp.loc[:, '1960':'1979'].columns, axis=1) 
df_gdp = df_gdp.drop('2019', axis=1)
```

```python
df_gdp['Total'] = df_gdp.sum(axis=1)
```

```python
#df_gdp.describe()
```

```python
df_gdp = df_gdp.rename(columns={'Country Name': 'Country'})
```

```python
df_gdp = df_gdp.set_index(['Country'])
```

```python
df_gdp = df_gdp.drop(['Arab World','Central Europe and the Baltics','East Asia & Pacific (excluding high income)',
             'Early-demographic dividend','East Asia & Pacific','Europe & Central Asia (excluding high income)',
             'Europe & Central Asia','Euro area','European Union','Fragile and conflict affected situations','High income',
             'Heavily indebted poor countries (HIPC)','IBRD only','IDA & IBRD total','IDA total','IDA blend','IDA only',
             'Latin America & Caribbean (excluding high income)','Lao PDR','Latin America & Caribbean',
             'Least developed countries: UN classification','Low income','Lower middle income','Low & middle income',
             'Late-demographic dividend','Middle East & North Africa','Middle income','Middle East & North Africa (excluding high income)',
             'OECD members','Other small states','Pre-demographic dividend','West Bank and Gaza','Pacific island small states',
             'Post-demographic dividend','South Asia','Sub-Saharan Africa (excluding high income)','Small states',
             'Turks and Caicos Islands','East Asia & Pacific (IDA & IBRD countries)','Europe & Central Asia (IDA & IBRD countries)',
             'Latin America & the Caribbean (IDA & IBRD countries)','Middle East & North Africa (IDA & IBRD countries)',
             'South Asia (IDA & IBRD)','Sub-Saharan Africa (IDA & IBRD countries)','Upper middle income','World','North America'], 
            axis=0)
#df_gdp.head()
```

```python
for country_name in ["United States", "China", "Brazil"]:
    df_country_name = df_gdp.loc[country_name, years]
    df_country_name.plot()
    plt.title(country_name)
    plt.ylabel('GDP(current US$)')
    plt.xlabel('Years')
    plt.show()
```

```python
United_States.plot(kind='line')

plt.title('GDP of United States')
plt.ylabel('GDP(current US$)')
plt.xlabel('Years')

# annotate the 2007-2008 financial crisis. 
# syntax: plt.text(x, y, label)
plt.text(2007,1.31e13, '2008 financial crisis', color='blue')

plt.show() 
```

```python
gdp_China_US = df_gdp.loc[['China','United States'], years] 
#gdp_China_US.head()
```

```python
gdp_China_US=gdp_China_US.transpose()
#gdp_China_US.head()
```

```python
gdp_China_US.index = gdp_China_US.index.map(int)
gdp_China_US.plot(kind='line')

plt.title('GDP of China and United States')
plt.ylabel('GDP(current US$)')
plt.xlabel('Years')

plt.show()
```

```python
df_gdp = df_gdp.sort_values(by='Total',ascending=False,axis=0)
```

```python

df_gdp_top5=df_gdp.head(5)

df_gdp_top5=df_gdp_top5[years].transpose()

#print(df_gdp_top5)
```

```python
df_gdp_top5.index=df_gdp_top5.index.map(int)
df_gdp_top5.plot(kind='line', figsize=(16,10))
plt.title('GDP of Top 5 Countries')
plt.ylabel('GDP(current US$)')
plt.xlabel('Years')
plt.show
```

```python
df_gdp_top5.index=df_gdp_top5.index.map(int)
df_gdp_top5.plot(kind='area', 
                 stacked=False,
                 figsize=(20, 10),
                 )
plt.title('GDP of Top 5 Countries')
plt.ylabel('GDP(current US$)')
plt.xlabel('Years')
plt.show
```

```python
df_gdp_top5.index=df_gdp_top5.index.map(int)
df_gdp_top5.plot(kind='area',
                stacked=False,
                alpha=0.3,
                figsize=(20,10),
                )
plt.title('GDP of Top 5 Countries')
plt.ylabel('GDP(current US$)')
plt.xlabel('Years')
plt.show
```

```python
ax = df_gdp_top5.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('GDP of Top 5 Countries')
ax.set_ylabel('GDP(current US$)')
ax.set_xlabel('Years')
```
